### Data Exploration:

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('HRDataset_v14.csv')
df
```

### Finding the characteristics of dataset

```python
df.describe()
```

### Finding the employees with high performance

```python
# Employees who either 'fully meets' or 'exceeds' their performance are found first.
df1 = df[df.PerformanceScore != 'Needs Improvement']
df2 = df1[df1.PerformanceScore != 'PIP']
df2
```

```python
# Then, employees with highest employee satisfaction are found.
df3 = df2[df2.EmpSatisfaction >= 4]
df3
```

```python
# from .describe function, we know that mean absence is 	10.237942
```

```python
# applying mean absences to df3 to find out punctual employees
df4 = df3[df3.Absences <= 10]
df4
```

```python
# finding employees who arrive late
df.DaysLateLast30.mean()
```

```python
# as this value is negligible, we will not be using it for sorting.
```

```python
# deleting data about employees who left their jobs by using opposite of .dropna() method
df5 = df4[df4['DateofTermination'].isnull()]
df5
```

```python
# resetting index numbers
df5.reset_index(drop=True, inplace=True)
df5
```

```python
df5.plot()
```

```python
df5.to_excel('Cleaned HR data.xlsx')
```
