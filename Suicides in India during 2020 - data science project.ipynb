# Creating an analysis of Suicide Cases with respect to Causes, Age and Gender in the year 2020. 
## Data obtained from OGD- Open Government Data (data.gov.in)
### file name : NCRB_ADSI-2020_TAble_2.0.csv
### Changed file name for simplicity = "2020 Suicides data.csv"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing data from file.
suicides_2020 = pd.read_csv("2020 Suicides data.csv")

#clearing the bottom column to avoid total value's inclusion to the plots.
suicides_2020.drop([30],axis=0,inplace=True)

#making proper index column
suicides_2020.drop('Si. No', axis = 1, inplace= True)
suicides_2020.index.rename('index', inplace=True)

suicides_2020
```

## making comparative graphs

```python
#setting the style
plt.style.use("seaborn-paper")

# subplot of male and female
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(50,10),
                              sharex=True)

#adding data t ax0
scatter = ax0.scatter(x=below_18["Cause"],
                     y=below_18["Below 18 years - Male"],
                     
                     cmap="summer") #this changes the colour scheme

#customizing ax0
ax0.set(title="Deaths in male below 18",
       xlabel="Causes",
       ylabel="Deaths")



#adding a meanline
ax0.axhline(y=below_18["Below 18 years - Male"].mean(),
           linestyle="--");



# adding data to ax1
scatter = ax1.scatter(x=below_18["Cause"],
                     y=below_18["Below 18 years - Female"],
                     
                     cmap="summer")

#customizing ax1
ax1.set(title = "Deaths in below 18 - female",
       ylabel = "Deaths")


#adding meanline
ax1.axhline(y=below_18["Below 18 years - Female"].mean(),
           linestyle="--");

#not plotting transgender table as those values are negligible.

#adding a title to a figure
fig.suptitle("Deaths in Male, Female, Transgenders below 18", fontsize=16, fontweight="bold");
```
